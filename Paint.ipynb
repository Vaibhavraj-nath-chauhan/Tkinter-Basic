```python
from tkinter import *
import tkinter.ttk as ttk
from tkinter import colorchooser
from tkinter import filedialog
from PIL import Image, ImageDraw, ImageGrab, ImageTk
import PIL
from tkinter import messagebox
```

```python

root = Tk()
root.title('Codemy.com')
root.geometry('800x600')



brush_color = 'black'
def paint(e):
    
    #brush paramaters
    brush_width = int(my_slider.get())
    #brush_color = 'green'
    #Brush type : BUTT, ROUND, PROJECTING
    brush_type_2 = brush_type.get()
    
    #start position
    x1 = e.x -1
    y1 = e.y -1

    #End position
    x2  = e.x +1
    y2 = e.y + 1
    
    #Drawing line
    my_canvas.create_line(x1,y1,x2,y2,fill=brush_color,width=brush_width,capstyle=brush_type_2,smooth = True)
    
#Change Sire of brush
def change_brush_size(e):
    slider_lable.config(text=int(my_slider.get() ))


def change_brush_color():
    global brush_color
    brush_color = "black"
    brush_color = colorchooser.askcolor(color=brush_color)[1]
    
    
def change_canvas_color():
    bg_color = "black"
    bg_color = colorchooser.askcolor(color=bg_color)[1]
    my_canvas.config(bg = bg_color)

#cleaar Screen
def clear_screen():
    my_canvas.delete(ALL)
    my_canvas.config(bg="white")
#Saving image
def save_image():
    result = filedialog.asksaveasfilename(initialdir="D:\\Learning\\College Section\\Language practice\\Python Practice\\Tkinker\\Udemy",filetypes=(("*png files","*.png"),("all files","*.*")))
    if result.endswith('.png'):
        pass
    else:
        result = result +".png"
    
    if result:
        x = root.winfo_rootx() + my_canvas.winfo_x()
        y = root.winfo_rooty() + my_canvas.winfo_y()
        x1 = x+ my_canvas.winfo_width() + 250
        y1 = y+ my_canvas.winfo_height() + 100
        ImageGrab.grab().crop((x,y,x1,y1)).save(result)
        #Pop up mesgse
        messagebox.showinfo("Image Saved", "Your Image is saved")
    

#creating Canvas
my_canvas = Canvas(root,width=400,height=300,bg="white")
my_canvas.pack(pady=20)


#my_canvas.create_line(x1,y1,x2,y2,fill='red')
#my_canvas.create_line(150,0,150,200,fill='red')

my_canvas.bind('<B1-Motion>',paint)


#Creating brush paramater frame
brush_edit_frame = Frame(root)
brush_edit_frame.pack(pady=20)

#Brush Size
brush_size_frame = LabelFrame(brush_edit_frame,text="Brush Size")
brush_size_frame.grid(row=0,column=0,padx=50)
#Brush Slider
my_slider = ttk.Scale(brush_size_frame,from_=1,to=100,command=change_brush_size,orient=VERTICAL,value=10)
my_slider.pack(pady=10,padx=10)
#Brush slider label
slider_lable = Label(brush_size_frame,text=my_slider.get())
slider_lable.pack(pady=5)



#Brush Type
brush_type_frame = LabelFrame(brush_edit_frame,text="Brush Type")
brush_type_frame.grid(row=0,column =1,padx=50)

brush_type = StringVar()
brush_type.set('round')

#radio button on brust tupe
brush_type_radio1 = Radiobutton(brush_type_frame,text="Round",variable=brush_type, value="round")
brush_type_radio2 = Radiobutton(brush_type_frame,text="Slash",variable=brush_type, value="butt")
brush_type_radio3 = Radiobutton(brush_type_frame,text="Diamond",variable=brush_type, value="projecting")

brush_type_radio1.pack(anchor="w")
brush_type_radio2.pack(anchor="w")
brush_type_radio3.pack(anchor="w")



#Change Colors
change_colors_frame = LabelFrame(brush_edit_frame,text="Change Colors")
change_colors_frame.grid(row=0,column=2)

#change brush color button
brush_color_button = Button(change_colors_frame,text="Brush Color",command=change_brush_color)
brush_color_button.pack(pady=10,padx=10)

#change canvas background color
canvas_color_button = Button(change_colors_frame,text="Canvas Color",command=change_canvas_color)
canvas_color_button.pack(pady=10,padx=10)

#program Option 
option_frame = LabelFrame(brush_edit_frame,text="Program Option")
option_frame.grid(row=0,column=3,padx=50)

clear_button = Button(option_frame,text="Clear Screen", command=clear_screen)
clear_button.pack(padx=10,pady=10)

#save image
save_button = Button(option_frame,text="Save as PNG", command=save_image)
save_button.pack(padx=10,pady=10)


root.mainloop()
```
